# Student Loan Risk with Deep Learning

```python
# Imports
import pandas as pd
import tensorflow as tf
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report
from pathlib import Path
```

## Prepare the data to be used on a neural network model

### Step 1: Read the `student-loans.csv` file into a Pandas DataFrame. Review the DataFrame, looking for columns that could eventually define your features and target variables.   

```python
# Read the csv into a Pandas DataFrame
file_path = "https://static.bc-edx.com/ai/ail-v-1-0/m18/lms/datasets/student-loans.csv"
loans_df = pd.read_csv(file_path)

# Review the DataFrame
loans_df.head()
```

```python
# Review the data types associated with the columns
loans_df.dtypes
```

```python
# Check the credit_ranking value counts
loans_df["credit_ranking"].value_counts()
```

### Step 2: Using the preprocessed data, create the features (`X`) and target (`y`) datasets. The target dataset should be defined by the preprocessed DataFrame column “credit_ranking”. The remaining columns should define the features dataset.

```python
# Define the target set y using the credit_ranking column
y = loans_df["credit_ranking"]

# Display a sample of y
y[:5]

# Display a sample of y
y.head()
```

```python
# Define features set X by selecting all columns but credit_ranking
X = loans_df.drop(columns=["credit_ranking"])

# Review the features DataFrame
X.head()

```

### Step 3: Split the features and target sets into training and testing datasets.


```python
# Split the preprocessed data into a training and testing dataset
# Assign the function a random_state equal to 1
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)

```

### Step 4: Use scikit-learn's `StandardScaler` to scale the features data.

```python
# Create a StandardScaler instance
scaler = StandardScaler()


# Fit the scaler to the features training dataset
X_scaler = scaler.fit(X_train)


# Fit the scaler to the features training dataset
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

---

## Compile and Evaluate a Model Using a Neural Network

### Step 1: Create a deep neural network by assigning the number of input features, the number of layers, and the number of neurons on each layer using Tensorflow’s Keras.

> **Hint** You can start with a two-layer deep neural network model that uses the `relu` activation function for both layers.


```python
# Define the the number of inputs (features) to the model
number_input_features = len(X_train.iloc[0])

# Review the number of features
number_input_features
```

```python
# Define the number of hidden nodes for the first hidden layer
hidden_nodes_layer1 = (number_input_features + 1) // 2

# Review the number hidden nodes in the first layer
hidden_nodes_layer1


# Define the number of hidden nodes for the second hidden layer
hidden_nodes_layer2 = (hidden_nodes_layer1 + 1) // 2

# Review the number hidden nodes in the second layer
hidden_nodes_layer2


# Define the number of neurons in the output layer
number_output_neurons = 1
```

```python
# Create the Sequential model instance
nn = Sequential()


# Add the first hidden layer
nn.add(Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation="relu"))


# Add the second hidden layer
nn.add(Dense(units=hidden_nodes_layer2, activation="relu"))


# Add the output layer to the model specifying the number of output neurons and activation function
nn.add(Dense(units=number_output_neurons, activation="linear"))

```

```python
# Display the Sequential model summary
nn.summary()
```

### Step 2: Compile and fit the model using the `binary_crossentropy` loss function, the `adam` optimizer, and the `accuracy` evaluation metric.


```python
# Compile the Sequential model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Fit the model using 50 epochs and the training data
fit_model = nn.fit(X_train_scaled, y_train, epochs=50)
```

### Step 3: Evaluate the model using the test data to determine the model’s loss and accuracy.


```python
# Evaluate the model loss and accuracy metrics using the evaluate method and the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)

# Display the model loss and accuracy results
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")

```

### Step 4: Save and export your model to a keras file, and name the file `student_loans.keras`.


```python
# Set the model's file path
file_path = Path("student_loans.keras")


# Export your model to a keras file
nn.save(file_path)
```

## Predict Loan Repayment Success by Using your Neural Network Model

### Step 1: Reload your saved model.

```python
# Set the model's file path
file_path = Path("student_loans.keras")


# Load the model to a new object
nn_imported = tf.keras.models.load_model(file_path)
```

### Step 2: Make predictions on the testing data and save the predictions to a DataFrame.

```python
# Make predictions with the test data
predictions = nn_imported.predict(X_test_scaled).round().astype("int32")


# Display a sample of the predictions
display(predictions)
```

```python
# Save the predictions to a DataFrame and round the predictions to binary results
predictions_df = pd.DataFrame(predictions, columns=["Predictions"])

```

### Step 4: Display a classification report with the y test data and predictions

```python
# Print the classification report with the y test data and predictions
print(classification_report(y_test, predictions))
```

1. To build a recommendation system for student loans, you would need to collect a variety of personal and financial data from students, along with information about available loan options. Relevant data may include:

Student Information:

Demographics: Age, marital status, dependents, state of residence, citizenship status (as some loans may only be available to residents or citizens).
Academic Information: Degree type (e.g., undergraduate, graduate), major, expected graduation date, current year in school, part-time/full-time status.
Financial Information:
Current Income/Employment: This can help determine if students are eligible for certain loans that require income verification.
Credit Score: Many private loans require creditworthiness to assess eligibility and interest rates.
Financial Aid: Previous financial aid packages, current scholarships, or grants.
Cost of Attendance: Tuition fees, living expenses, and other related expenses for the specific school.
Loan Preferences: Loan term preferences (short-term vs long-term loans), preferred repayment plans (fixed interest, variable interest), and any co-signer availability.
Loan Information:

Loan Type: Government loans (subsidized/unsubsidized), private loans, refinancing options.
Interest Rates: Fixed or variable rates, loan amounts, and available repayment plans.
Eligibility Criteria: Based on credit scores, cosigner requirements, income verification, etc.
This data is relevant and appropriate because student loan options often depend on the specific financial situation, credit history, and educational details of each student. The more detailed the data you collect, the more personalized and accurate the recommendations can be.

2. Given the data chosen, the most suitable filtering method for this system would likely be a content-based filtering approach. Here’s why:

Content-Based Filtering: This method makes recommendations based on the attributes of the loan products and the specific features of the student (e.g., income, credit score, loan amount needed). Since the system will match students with loans that fit their specific financial and academic profiles, the recommendation process is based on the attributes (or "content") of both the student and the available loans.

Why Content-Based Filtering:

The student’s financial details, academic background, and loan preferences provide rich information to compare against the characteristics of different loan options.
For example, if a student has a strong credit score and needs a loan with a low-interest rate, content-based filtering can match them with loans that fit these criteria.

3. Challenge 1: Privacy and Data Security
Why It’s a Concern: Student loan recommendations require highly sensitive personal and financial information, such as income, credit scores, and loan preferences. This type of data is vulnerable to breaches, and improper handling could lead to identity theft or financial fraud. Therefore, ensuring robust data encryption, secure storage, and strict privacy policies are crucial.
Mitigation: The recommendation system would need to comply with relevant privacy regulations (e.g., GDPR, FERPA in the US) and use advanced security measures like encryption, secure access control, and anonymization techniques when handling data.
Challenge 2: Loan Suitability and Fairness
Why It’s a Concern: Different students come from different financial backgrounds, and recommending unsuitable loans (e.g., high-interest loans to low-income students) could lead to financial burdens. Ensuring that the system provides fair and equitable recommendations to all students—especially those from disadvantaged backgrounds—is crucial.
Mitigation: The recommendation system should incorporate a fairness mechanism that considers the long-term affordability of loans, promotes responsible borrowing, and avoids bias (e.g., disproportionately offering better terms to students with higher incomes). Transparency in the recommendation process is also important, so students understand why certain loans are recommended.
